```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
```

Reading the Data Set

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/teleCust1000t.csv')
df.head()
```

```python
df['custcat'].value_counts()
```

```python
df.columns
```

```python
X = df[['region', 'tenure', 'age', 'marital', 'address', 'income', 'ed',
       'employ', 'retire', 'gender', 'reside']]
X[0:5]
```

```python
Y = df['custcat']
Y[0:5]
```

```python
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=4)
print(X_train.shape,X_test.shape)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=4
neig = KNeighborsClassifier(n_neighbors=k).fit(X_train,Y_train)
neig
```

```python
yhat=neig.predict(X_test)
yhat[0:5]
```

Finding Accuarcy

```python
from sklearn import metrics
print("Train_accuarcy:",metrics.accuracy_score(Y_train,neig.predict(X_train)))
print("Test_accuracy:",metrics.accuracy_score(Y_test,yhat))
```
